## Producer Code

```python
def read_csv(fileName):
    '''Read the CSV file clickstream.csv'''
    data = []
    #WRITE THE CODE TO READ CSV FILE
    with open(fileName, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
            
    return data         

 
def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, value=data)
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'Streaming_Linux_process'
    
    cRows = read_csv('Streaming_Linux_process.csv')
    
    Val4 = ['4']
    mac4 = list(filter(lambda i: i['machine'] in Val4, cRows))

    Val5 = ['5']
    mac5 = list(filter(lambda i: i['machine'] in Val5, cRows))

    Val6 = ['6']
    mac6 = list(filter(lambda i: i['machine'] in Val6, cRows))

    Val7 = ['7']
    mac7 = list(filter(lambda i: i['machine'] in Val7, cRows))

    Val8 = ['8']
    mac8 = list(filter(lambda i: i['machine'] in Val8, cRows))
    
    print('Publishing records..')
    producer = connect_kafka_producer()

    len_mac4 = len(mac4)
    len_mac5 = len(mac5) 
    len_mac6 = len(mac6) 
    len_mac7 = len(mac7) 
    len_mac8 = len(mac8) 
    # defining starting and ending index
    start_ind4 = 0
    end_ind4 = 0
    start_ind5 = 0
    end_ind5 = 0
    start_ind6 = 0
    end_ind6 = 0
    start_ind7 = 0
    end_ind7 = 0
    start_ind8 = 0
    end_ind8 = 0
    
    while True:
        #Machine 4
        rand0m4 = random.randint(10,50)
        if len_mac4>rand0m4:
            start_ind4 = end_ind4
            end_ind4 = start_ind4+rand0m4
            data4 = mac4[start_ind4:end_ind4]
            len_mac4 = len_mac4-rand0m4
            
            
        else:
            rand0m4 = len_mac4
            start_ind4 = end_ind4
            end_ind4 = start_ind4+rand0m4
            data4 = mac4[start_ind4:end_ind4]
            start_ind4 = 0
            end_ind4 = 0
            len_mac4 = len(mac4)
            
        
        for k in data4:
            k['ts'] = int(dt.datetime.timestamp(dt.datetime.now(timezone('UTC'))))#appending unix timestamp

        
        
        # Machine 5
        rand0m5 = random.randint(10,50)
        if len_mac5>rand0m5:
            start_ind5 = end_ind5
            end_ind5 = start_ind5+rand0m5
            data5 = mac5[start_ind5:end_ind5]
            len_mac5 = len_mac5-rand0m5
            
        else:
            rand0m5 = len_mac5
            start_ind5 = end_ind5
            end_ind5 = start_ind5+rand0m5
            data5 = mac5[start_ind5:end_ind5]
            start_ind5 = 0
            end_ind5 = 0
            len_mac5 = len(mac5)
        
        for k in data5:
            k['ts'] = int(dt.datetime.timestamp(dt.datetime.now(timezone('UTC')))) #appending unix timestamp

        
        
        # Machine 6
        rand0m6 = random.randint(10,50)
        if len_mac6>rand0m6:
            start_ind6 = end_ind6
            end_ind6 = start_ind6+rand0m6
            data6 = mac6[start_ind6:end_ind6]
            len_mac6 = len_mac6-rand0m6
            
        else:
            rand0m6 = len_mac6
            start_ind6 = end_ind6
            end_ind6 = start_ind6+rand0m6
            data6 = mac6[start_ind6:end_ind6]
            start_ind6 = 0
            end_ind6 = 0
            len_mac6 = len(mac6)
        
        for k in data6:
            k['ts'] = int(dt.datetime.timestamp(dt.datetime.now(timezone('UTC')))) #appending unix timestamp

        
        
        # Machine 7
        rand0m7 = random.randint(10,50)
        if len_mac7>rand0m7:
            start_ind7 = end_ind7
            end_ind7 = start_ind7+rand0m7
            data7 = mac7[start_ind7:end_ind7]
            len_mac7 = len_mac7-rand0m7
            
        else:
            rand0m7 = len_mac7
            start_ind7 = end_ind7
            end_ind7 = start_ind7+rand0m7
            data7 = mac7[start_ind7:end_ind7]
            start_ind7 = 0
            end_ind7 = 0
            len_mac7 = len(mac7)
        
        for k in data7:
            k['ts'] = int(dt.datetime.timestamp(dt.datetime.now(timezone('UTC')))) #appending unix timestamp

        
        
        
        # Machine 8
        rand0m8 = random.randint(10,50)
        if len_mac8>rand0m8:
            start_ind8 = end_ind8
            end_ind8 = start_ind8+rand0m8
            data8 = mac8[start_ind8:end_ind8]
            len_mac8 = len_mac8-rand0m8
            
        else:
            rand0m8 = len_mac8
            start_ind8 = end_ind8
            end_ind8 = start_ind8+rand0m8
            data8 = mac8[start_ind8:end_ind8]
            start_ind8 = 0
            end_ind8 = 0
            len_mac8 = len(mac8)
        
        for k in data8:
            k['ts'] = int(dt.datetime.timestamp(dt.datetime.now(timezone('UTC')))) #appending unix timestamp

        
        
        combined_data = data4+data5+data6+data7+data8
        publish_message(producer, topic, combined_data)
        sleep(5)
```
